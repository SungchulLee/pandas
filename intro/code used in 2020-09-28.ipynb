```python
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df.head(3)
```

```python
print(df.iloc[1, 3]) # implicit index
print(df.loc[1, 'Name']) # explict index
```

```python
print(df.iloc[1:3, 3:5]) # implicit index 스라이싱 (끝점 미포함)
print(df.loc[1:3, 'Name':'Age']) # explicit index 스라이싱 (끝점 포함!!!!)
```

```python
print(df.iloc[1, 3].shape) # implicit index
print(df.loc[1, 'Name'].shape) # explict index
```

```python
print(df.iloc[1:2, 3].shape) # implicit index
print(df.loc[1:1, 'Name':'Name'].shape) # explict index
```

```python
import yfinance as yf
tickers = ['FB', 'AMZN', 'AAPL', 'NFLX', 'GOOGL']
for ticker in tickers:
    stock = yf.Ticker(ticker)
    df = stock.history(period='max')
    df.to_csv(f'data/{ticker}.csv')
```
