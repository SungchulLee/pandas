```python
import pandas as pd

df = pd.DataFrame()

tickers = ['FB', 'AMZN', 'AAPL', 'NFLX', 'GOOGL']
for ticker in tickers:
    df_ = pd.read_csv(f'data/{ticker}.csv',
                      usecols=['Close','Date'],
                      index_col='Date',
                      parse_dates=True).rename(columns={"Close": ticker})
    df = df.join(df_, how='outer')

df.head()
```
